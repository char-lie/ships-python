```python
%matplotlib inline

from matplotlib.colors import LogNorm
from matplotlib.cm import Greys
from matplotlib import pyplot
from numpy import load

Z = load("12.npy")
density = Z[Z.shape[0] // 4:3 * Z.shape[0] // 4, 2 * Z.shape[1] // 5:Z.shape[1] // 2]
density[density < 1E-6] = 0

interesting_probabilities = (1E-6, 1E-5, 5E-5)
for probability_cut in interesting_probabilities:
    cumulative_probability = Z[Z > probability_cut].sum()
    print(
        "Sum of points with probability greater than "
        f"{probability_cut:.1e} "
        f"is {cumulative_probability * 100:<2.2f}%"
    )

pyplot.clf()
fig, ax = pyplot.subplots(figsize=(5, 10))
ax.set_title("Distribution after 12 hours")

extent = [300, 350, -50, 50]
pyplot.grid()
img = ax.imshow(
    density,
    norm=LogNorm(),
    cmap=Greys,
    extent=extent,
)
image_colorbar = pyplot.colorbar(img, orientation='horizontal')

contour = pyplot.contour(density, levels=interesting_probabilities, extent=extent)
contour_colorbar = pyplot.colorbar(contour)
contour_colorbar.lines[0].set_linewidth(200)
```
